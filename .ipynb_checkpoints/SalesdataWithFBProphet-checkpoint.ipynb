```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from fbprophet import Prophet
from datetime import datetime
from datetime import timedelta
% matplotlib inline
```

```python
df = pd.read_csv('sales_train.csv')
```

```python
df.head()
```

```python
df.shape
```

Let's parse the date column (Date format is DD.MM.YYYY (string))
Important: pd.to_datetime is extremely slow if we don't pass explicit format argument (format='%d.%m.%Y')

```python
df['date'] = pd.to_datetime(df['date'], format='%d.%m.%Y')
```

```python
df.head(n=3)
```

OK, let's start with item for there is most data:

```python
fav = df['item_id'].mode()[0]
print('Most popular item_id: ' + str(fav))
print('Number of occurrences: ' + str((df['item_id']==fav).sum()))
```

And take the shop which sells the most of this item:

```python
fav_shop_item_pair = df.loc[ df['item_id']==fav ]['shop_id'].mode()[0]
X = df.loc[ (df['item_id']==fav) & (df['shop_id']==fav_shop_item_pair) ]
print(X.shape)
```

```python
X.head()
```

```python
X = X.sort_values(by=['date'])
```

```python
X.plot(x='date', y='item_cnt_day', style='.-', figsize=(18,5))
```

There seems to be a gap in the data for 2015?!

```python
[l, r] = [pd.to_datetime(d) for d in ['2015-02', '2015-05']]
X.loc[ (X['date']>l) & (X['date']<r) ].plot(x='date', y='item_cnt_day', style='.-')
```

Let's look at gaps in the series using .diff  (by default compares each element to the one above it) and pd.value_counts():

```python
pd.value_counts(X['date'].diff())
```

```python
pd.value_counts(X['item_price'])
```

TIME SERIES MODELING USING fbprophet:

Instantiate the Prophet object and fit it to the (historical) dataframe:

```python
X_ts = Prophet()
X_ts.fit(X[['date', 'item_cnt_day']].rename(columns={"date": "ds", "item_cnt_day": "y"}))
```

Extend dataframe to "future" values (6months) for forecasting:

```python
X_future = X_ts.make_future_dataframe(periods=182)
X_future.tail(n=5)
```

Produce our initial forecast (all parameters are set to default)...

```python
X_forecast = X_ts.predict(X_future)
```

```python
Xfig = X_ts.plot(X_forecast)
Xfig.set_figwidth(15)
Xfig_ax=plt.gca()
#Xfig_ax.set_xlim(pd.Timestamp('2013-10'), pd.Timestamp('2016-01'))
```

Zoom in a bit:

```python
Xfig = X_ts.plot(X_forecast)
Xfig.set_figwidth(15)
Xfig_ax=plt.gca()
Xfig_ax.set_xlim(pd.Timestamp('2014-09'), pd.Timestamp('2016-03'))
```

There are several issues with this initial model, eg:
1. The model predicts negative values
2. It is strange that the large gap in early 2015 is fitted as a straight line.. but maybe this is intentional?

We can also plot the components separately

```python
Xcomps = X_ts.plot_components(X_forecast)
```

Try logistic instead of linear growth for the trend component:

```python
[flr, cap] = [0, 60]
X['floor'] = flr
X['cap'] = cap
X_future['cap'] = cap
X_future['floor'] = flr

X_ts = Prophet(growth='logistic')
X_ts.fit(X[['date', 'item_cnt_day', 'floor', 'cap']].rename(columns={"date": "ds", "item_cnt_day": "y", "floor": "floor", "cap": "cap"}))
X_forecast = X_ts.predict(X_future)
Xfig = X_ts.plot(X_forecast)
Xfig.set_figwidth(15)
Xfig_ax=plt.gca()
```

```python
Xcomps = X_ts.plot_components(X_forecast)
```
